```python
from playsound import playsound
import time
from IPython.display import Audio, display
```

```python
sound_file='/content/alarmsesi.mp3'
```

```python
CLEAR = "\033[2J"
CLEAR_AND_RETURN = "\033["
```

```python
def alarm(saniyeler):
  gecen_sure = 0

  print(CLEAR)
  while gecen_sure < saniyeler:
    time.sleep(1)
    gecen_sure += 1

    kalan_sure = saniyeler - gecen_sure
    kalan_dakika = kalan_sure // 60
    kalan_saniye = kalan_sure % 60 # 128 % 60 = 8

    print(f"{CLEAR_AND_RETURN} alarm çalacak: {kalan_dakika:02d}:{kalan_saniye:02d}")
  display(Audio(sound_file, autoplay=True))
dakikalar=int(input("ne kadar dakika beklenmeli?: "))
saniyeler=int(input("ne kadar saniye beklenmeli?: "))
toplam_saniye = dakikalar * 60 + saniyeler
alarm(10)
```
